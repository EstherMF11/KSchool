**Represent the logistic, or sigmoid, function between -20 and 20. Remember its formula is:**

$$  S(x)={\frac {1}{1+e^{-x}}}={\frac {e^{x}}{e^{x}+1}} $$



```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

#Creamos la función:
# sigmod = lambda(x) = 1 / (1+exp(-x)))
def sigmoid(x):
    """Returns the sigmoid function"""
    return (1/(1+np.exp(-x)))

# Linespace genera un rango desde el valor de inicio hasta el final
# Este ejemplo: la función va desde -20 hasta 20 con un total de x(10 numeros) e y (100numeros).

x = np.linspace(-20,20,10)
y = np.linspace(-20,20,100)

#Prepatación de plot, añadimos el color de cada una ('x' e 'y') y el label
 
plt.plot(x, sigmoid(x), 'r', label='linspace(-20,20,20)')
plt.plot(y, sigmoid(y), 'b', label='linspace(-20,20,100)')

# DIbujar línea de cuadrícula
plt.grid()

# Titulo y subtitulo
plt.title('Sigmoid Function')
plt.suptitle('Sigmoid')

# Nombre de los ejes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Poner la leyenda alineada a la derecha
plt.legend(loc='lower right')

# Escribir un texto con la fórmula
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)

# Mostrar la gráfica
plt.show()
```
