**1. Create a sample of points that follow the equation  ùëå=ùê¥ùëã+ùêµ , where A = 2.5 and B = 20. <br>
Now, plot it as either a cloud of points or a line.**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


#random xs data 
xs = np.random.normal(size = 100)

#Constants
a=2.5
b=20

#Creating lineal equation Y=AX+B
ys = a*xs + b

#Plotting 
plt.plot(xs,ys, '-o') 
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.title('First regression plot')



```
