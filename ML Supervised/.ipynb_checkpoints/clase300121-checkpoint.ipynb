# Basic Car Speed 

## Problem Statement

We would like to determine the speed of a car based on the bumpiness and slope of the road. For this, we have a list of the speed from a list of drivers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
%pylab inline
import pandas as pd
```

```python
df = pd.read_csv("terrain.csv")
```

```python
X = df[["bumpiness","grade"]].values
y = df["target"]
```

```python
def draw(clf,X,y,h=0.05):
    
    plt.figure(figsize=(10,10))

    x_min, x_max = X[:, 0].min() - 0.05, X[:, 0].max() + .05
    y_min, y_max = X[:, 1].min() - .05, X[:, 1].max() + .05
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    if clf is not None:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.scatter(xx,yy,c=Z)
    else:
        plt.scatter(xx,yy)
    
    plt.scatter(X[:,0],X[:,1],c=y,cmap="Paired")
```

```python
draw(None,X,y,0.01)
```

#### Regresin Logstica

```python
import sklearn
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X, y)
```

```python
draw(clf, X, y, 0.01)
```

```python
#Cross val score
from sklearn.model_selection import cross_val_score
cross_val_score(clf,X,y,scoring="accuracy")
```

```python
cross_val_score(clf,X,y,scoring="accuracy").mean()
```

#### K nearst neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Encontramos la mejor k
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
clf_gk = GridSearchCV(KNeighborsClassifier(),
                        param_grid={'n_neighbors':np.arange(1,100)},
                        scoring="accuracy")
#El rango debe ir de 1 a 100, no puede tener 0 vecinos
clf_gk.fit(X,y)
clf_gk.best_params_
```

```python
draw(clf_gk.best_estimator_, X, y, 0.01)
```

Vemos que abajo no hace un corte diagonal, la tendencia es horizontal. La forma de mejorarlo es tener ms datos.

```python
cross_val_score(clf_gk.best_estimator_, X, y, cv=5, scoring='accuracy').mean()
```

## Metricas

#### Precision
Precision - Prediction
$$Precision=\frac{True Positive}{Actual Results} or \frac{True Positive}{True Positive + False Positive}$$
Fast -> $\frac{Pred Fast \cap Real Fast}{Pred Fast}$

#### Recall
Recall - Real
$$Recall=\frac{True Positive}{Predicted Results} or \frac{True Positive}{True Positive + False Positive}$$
Fast -> $\frac{Pred Fast \cap Pred Fast}{Real Fast}$
$$Precision=\frac{True Positive}{True Positive + False Positive}$$

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

#### Support Vector Machine

Las Mquinas de Vector Soporte se fundamentan en el concepto de hiperplano. Este hiperplano es un plano de dimensiones $p-1$, donde $p$ es el nmero de variables independientes, que trata de separar el espacio p-dimensional en dos mitades, dejando a cada lado del mismo, una categora diferente de los datos.

Dado que un espacio se puede dividir de mltiples formas, el hiperplano ptimo de separacin (maximal margin hyperplane) es aquel que se encuentra lo ms alejado de posible de las observaciones del conjunto de entrenamiento. Para obtener este hiperplano se calculan las distancias perpendiculares desde las observaciones ms cercanas al plano de separacin, estas distancias son denominadas vector soporte.

Existn casos que por la naturaleza de las observaciones, stas no son perfectamente separables por un maximal margin hyperplane. En estos casos se busca un hiperplano que cometa el nmero mnimo de errores, conocido como support vector classifier - SVC.

Cuando los grupos no son separables de forma lineal en el espacio origen, una estrategia es aumentar las dimensiones del espacio para buscar un hiperplano que s sea capaz de diferenciar los grupos. Los lmites de separacin lineales generados en el espacio aumentado se convierten en lmites de separacin no lineales al proyectarlos en el espacio original.

$$f(x_1, x_2) = (x_1, x_2, x_1^2x_2^2)$$
Es en estos casos, cuando se introducen dimensiones nuevas en los datos, que el support vector classifier - SVC, se pasa a denominar support vector machines - SVM.

Para aumentar y transformar las dimensiones de un espacio se pueden utlizar los kernel. Un kernel (K) es una funcin que devuelve el resultado del dot product entre dos vectores realizado en un nuevo espacio dimensional distinto al espacio original en el que se encuentran los vectores. A continuacin se listan los kernel ms utilizados:

Kernel lineal:

Si se emplea un Kernel lineal, el clasificador support vector machine obtenido es equivalente al support vector classifier.

$$K(x, x') = xx'$$
Kernel polinmico:

No suele ser recomendable emplear valores de $d$ mayores 5 por problemas de overfitting.

$$K(x, x') = (xx'+c)^d$$
Kernel gaussiano (RBF):

El valor de $\gamma$ controla el comportamiento del kernel, cuando es muy pequeo, el modelo final es equivalente al obtenido con un kernel lineal, a medida que aumenta su valor, tambin lo hace la flexibilidad del modelo.

$$K(x, x') = e^{-\gamma||x-x'||^2} $$
Principales parmetros del modelo:

C: parmetro que controla las violaciones del margen, cuanto ms elevado, ms permisivo. Permite controlar el equilibrio entre sesgo y varianza del modelo.
kernel:
* linear: linea de separacin
* poly: linea polinomial de separacin
* degree: grado del polinomio
* rbf: crculo de separcin
* gamma: inverso del radio
